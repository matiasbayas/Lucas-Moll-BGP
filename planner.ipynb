```python
import numpy as np
import matplotlib.pyplot as plt
import growthmethods as gm

np.seterr(divide='ignore', invalid='ignore')
```

```python
# Parameters
rho = 0.06
theta = 0.5
k = 0.05

# Meeting technology
eta = 0.3
alpha0 = 0.0849

# Updating parameter
xi = 0.4

# Grid
I = 2001
xmin = 0 
xmax = 3
```

```python
# Planner:
w, sigmap, alphap, fp, Fp, gammap, xgrid, itp = gm.KGAT_pl(rho, theta, k, eta, alpha0, xi, xmin, xmax, I)

# Equilibrium
v, sigma, alpha, f, F, gamma, xgridp, it = gm.KGAT(rho, theta, k, eta, alpha0, xi, xmin, xmax, I)
```

```python
# Need to normalize by a common factor since units are arbitrary
# Use median productivity as normalization factor.

med = xgrid[[0.499 < i < 0.501 for i in F]]
medp = xgridp[[0.499 < i < 0.501 for i in Fp]]
```

```python
index = xgrid <= med*11/10
index2 = xgridp <= medp*11/10
plt.plot(xgrid[index]/med, sigma[index])
plt.plot(xgrid[index2]/medp, sigmap[index2])

plt.title('Planner Policy Function vs. Equilibrium Policy Function')
plt.xlabel('Productivity - $x$ (relative to Median)')
plt.legend(['Equilibrium', 'Planner'])
plt.xlim([0.2,11/10])

plt.savefig('KGAT5.png')
```
